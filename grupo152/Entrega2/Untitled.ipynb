```python
comunas = []

with open("direcciones.csv") as direcciones:
    for line in direcciones.readlines()[1:]:
        line = line.strip().split(",")
        comunas.append(line[2])
        
with open("unidades.csv") as direcciones:
    for line in direcciones.readlines()[1:]:
        line = line.strip().split(",")
        comunas.append(line[3])
        
comunas = (list(dict.fromkeys(comunas)))

f = open("comunas.csv", "w")
f.write("nombre_comuna")
for comuna in comunas:
    f.write("\n")
    f.write(f"{comuna}")
f.close()
```

```python
unidades2 = []
unidades_comunas = []

with open("unidades.csv") as unidades:
    for line in unidades.readlines()[1:]:
        line = line.strip().split(",")
        unidades2.append(f"{','.join(line[0:3])}")
        unidades_comunas.append(f"{line[0]},{comunas.index(line[3])}")
        
f = open("unidades2.csv", "w")
f.write("uid,direccion,jefe")
for unidad in unidades2:
    f.write("\n")
    f.write(f"{unidad}")
f.close()

f = open("unidades_comunas.csv", "w")
f.write("uid,comid")
for unidad_comuna in unidades_comunas:
    f.write("\n")
    f.write(f"{unidad_comuna}")
f.close()
```

```python
direcciones2 = []

with open("direcciones.csv") as direcciones:
    for line in direcciones.readlines()[1:]:
        line = line.strip().split(",")
        direcciones2.append(f"{','.join(line[0:2])},{comunas.index(line[2])}")
        
f = open("direcciones2.csv", "w")
f.write("dirid,nombre,comid")
for direccion in direcciones2:
    f.write("\n")
    f.write(f"{direccion}")
f.close()        
```

```python
vehiculos2 = []
carga = []
cfria = []
pfrescos = []

with open("vehiculos.csv") as vehiculos:
    for line in vehiculos.readlines()[1:]:
        line = line.strip().split(",")
        vehiculos2.append(f"{line[0]},{line[-1]},{','.join(line[1:3])},{line[3].strip(' ')}")
        if line[4]:
            carga.append(f"{line[0]},{','.join(line[4:6])}")
        if line[6]:
            pfrescos.append(f"{line[0]},{line[6]}")
        if line[7]:
            cfria.append(f"{line[0]},{','.join(line[7:9])}")
                
f = open("vehiculos2.csv", "w")
f.write("vid,uid,patente,estado,tipo")
for vehiculo in vehiculos2:
    f.write("\n")
    f.write(f"{vehiculo}")
f.close()     

f = open("carga.csv", "w")
f.write("vid,volumen,carga")
for vehiculo in carga:
    f.write("\n")
    f.write(f"{vehiculo}")
f.close()     

f = open("pfrescos.csv", "w")
f.write("vid,alcance")
for vehiculo in pfrescos:
    f.write("\n")
    f.write(f"{vehiculo}")
f.close()     

f = open("cfria.csv", "w")
f.write("vid,cantidad_compartimientos,capacidad_compartimientos")
for vehiculo in cfria:
    f.write("\n")
    f.write(f"{vehiculo}")
f.close()     
```

```python
personal2 = []
administrativos = []
repartidores = []
repartidores_vehiculos = []

with open("personalV2.csv") as personal:
    for line in personal.readlines()[1:]:
        line = line.strip().split(",")
        personal2.append(f"{','.join(line[0:5])}")
        if line[6]:
            administrativos.append(f"{line[0]},{line[6]},{line[5]}")
        if line[7]:
            repartidores.append(f"{line[0]},{line[7]}")
            repartidores_vehiculos.append(f"{line[0]},{line[-1]}")
            
f = open("personal2.csv", "w")
f.write("pid,nombre,rut,sexo,edad")
for empleado in personal2:
    f.write("\n")
    f.write(f"{empleado}")
f.close()   

f = open("administrativos.csv", "w")
f.write("pid,uid,clasificacion")
for empleado in administrativos:
    f.write("\n")
    f.write(f"{empleado}")
f.close()  

repartidores = list(dict.fromkeys(repartidores))

f = open("repartidores.csv", "w")
f.write("pid,licencia")
for empleado in repartidores:
    f.write("\n")
    f.write(f"{empleado}")
f.close()  

f = open("repartidores_vehiculos.csv", "w")
f.write("pid,vid")
for empleado in repartidores_vehiculos:
    f.write("\n")
    f.write(f"{empleado}")
f.close()  
```
